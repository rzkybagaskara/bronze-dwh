### 4. EDA

Before we transform the data, first we want to take a look at the data count/distribution. We want to know the car and detail_iklan data based on the dataframes above.

We want to know which car models, transmission, types, and colours are listed the most in the used car data above.

```python
def plot_counts(dataframe, columns:str, plot: False):
    ordered = dataframe[columns].value_counts().index #function to sort the data by descending order
    if plot is True:
        plt.figure(figsize=(20,6))
        plt.title(f'{columns} counts')
        sns.countplot(dataframe, x=columns, order=ordered)
        plt.show()
    return ordered
```

1. car

```python
car_brands = plot_counts(mobil_data, 'Merek', True)
car_brands
```

```python
car_transmissions = plot_counts(mobil_data, 'Transmisi', True)
car_transmissions
```

```python
car_types = plot_counts(mobil_data, 'Jenis', True)
car_types
```

```python
car_colours = plot_counts(mobil_data, 'Warna', True)
car_colours
```

We also want to know which regions have the most listings of used cars

2. detail_iklan

```python
#top 10 data
top_10_regions = detail_iklan_data['Domisili'].value_counts().head(10).index

filtered_top10_regions = detail_iklan_data[detail_iklan_data['Domisili'].isin(top_10_regions)]

ads_regions = plot_counts(filtered_top10_regions, 'Domisili', True)
ads_regions
```
