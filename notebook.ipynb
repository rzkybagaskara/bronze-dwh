We want to know which car models, transmission, types, and colours are listed the most in the used car data above.

```python
def plot_counts(dataframe, columns:str, plot: False):
    """
    Function to create a countplot using Seaborn based on specific column on a pandas dataframe.
    """
    ordered = dataframe[columns].value_counts().index #function to sort the data by descending order
    if plot is True:
        plt.figure(figsize=(20,6))
        plt.title(f'{columns} counts')
        sns.countplot(dataframe, x=columns, order=ordered)
        plt.show()
    return ordered
```

### 5. Data Transformation

1. Alamat fields

```python
def address_standardization(data, column: str):
    """
    Function to standardized address data on a pandas dataframe.
    """
    return (
        data[column]
        .astype(str) #convert to string to be able to use str.replace
        .str.replace(r'\bGg\.?\s*', 'Gang ', regex=True)
        .str.replace(r'\bJl\.?\s*', 'Jalan ', regex=True)
        .str.replace(r'\s+', ' ', regex=True)   #remove double spaces
        .str.strip()
    )
```

```python
new_ads_address = address_standardization(detail_iklan_data, 'Alamat')
new_ads_address.head(10)
```

```python
#replace the old field with new one
detail_iklan_data['Alamat'] = new_ads_address
detail_iklan_data.head(10)
```

```python
new_users_address = address_standardization(pengguna_data, 'Alamat')
new_users_address.head(10)
```

```python
#replace the old field with new one
pengguna_data['Alamat'] = new_users_address
pengguna_data.head(10)
```

Insert the data into SQL tables

```python
from sqlalchemy import create_engine

#create database connection to postgresql database
engine = create_engine("postgresql://postgres:bagas123@localhost/dwh_bronze")
```

```python
def insert_data_to_sql(csv_data, table_name:str, connection, table_status:str, index_status:bool):
    """
    Function to insert data from CSV files into SQL Tables. 
    csv_data -> csv file name
    table_name -> SQL table name (string format)
    connection -> engine name
    table_status -> append to insert (string format)
    index_status -> index of the dataframe (bool format), by default use False
    """
    return(
        csv_data.to_sql(
            name=table_name,
            con=connection,
            if_exists=table_status,
            index=index_status
        )
    )
```

```python
insert_cars_data = insert_data_to_sql(cars_data, 'cars', engine, 'append', False)
insert_cars_data
```
